### 使用CV扫描脸部，读取数据点


```python
#文档说明：识别人脸判断疲劳，作者考虑从打哈欠频率和闭眼频率入手
#同时避免唱歌张大嘴巴、眼睛进沙等特殊情况，需要两个同时满足才触发警报（楽）

from scipy.spatial import distance as dist
import numpy as np
import dlib
import cv2
import mysql.connector
import time

# MySQL数据库配置
mysql_config = {
    'host': '192.168.137.207',
    'database': 'fatigueTest',
    'user': 'root',
    'password': 'tansome'
}

# 清空数据库表格
def clear_table(cursor):
    cursor.execute('TRUNCATE TABLE eye_mouth')

# 添加数据到数据库表格
def add_data(cursor, time_value, blink_value, yawn_value):
    insert_query = 'INSERT INTO eye_mouth (time, blink, yawn) VALUES (%s, %s, %s)'
    cursor.execute(insert_query, (time_value, blink_value, yawn_value))

# 连接到MySQL数据库
connection = mysql.connector.connect(**mysql_config)
cursor = connection.cursor()

# 清空数据库表格
clear_table(cursor)

#先自动开启ssh，启动摄像头
# SSH连接信息
hostname = '192.168.137.240'  # 树莓派的IP地址
port = 22  # SSH端口号，默认为22
username = 'wwb'  # 树莓派的用户名
password = 'pi'  # 树莓派用户的密码

# 创建SSH客户端对象
client = paramiko.SSHClient()
client.set_missing_host_key_policy(paramiko.AutoAddPolicy())

try:
    # 连接到树莓派
    client.connect(hostname, port, username, password)
    
    # 执行sudo motion命令
    stdin, stdout, stderr = client.exec_command('sudo motion')

    # 获取命令执行结果
    output = stdout.read().decode('utf-8')
    print(output)
    
    # 关闭SSH连接
    client.close()

except paramiko.AuthenticationException:
    print("认证失败，请检查用户名和密码。")

except paramiko.SSHException as ssh_err:
    print("SSH连接错误: ", ssh_err)

except paramiko.Exception as e:
    print("错误: ", e)

# 定义视频源的URL
url = "http://192.168.137.240:8081/" #树莓派2，用于面部抓取疲劳值
# url = "http://mirror.aarnet.edu.au/pub/TED-talks/911Mothers_2010W-480p.mp4"#测试视频用

# 创建视频捕获对象
cap = cv2.VideoCapture(url)

# 检查视频捕获对象是否成功打开
if not cap.isOpened():
    print("无法打开视频源")
    exit()

#定义脸上采集点
FACIAL_LANDMARKS_68_IDXS = dict([
    ("mouth", (48, 68)),
    ("right_eyebrow", (17, 22)),
    ("left_eyebrow", (22, 27)),
    ("right_eye", (36, 42)),
    ("left_eye", (42, 48)),
    ("nose", (27, 36)),
    ("jaw", (0, 17))
])

#眨眼计算函数
def eye_aspect_ratio(eye):
    # 计算距离，竖直的
    A = dist.euclidean(eye[1], eye[5])
    B = dist.euclidean(eye[2], eye[4])
    # 计算距离，水平的
    C = dist.euclidean(eye[0], eye[3])
    # ear值
    ear = (A + B) / (2.0 * C)
    return ear

#打哈欠计算函数
def mouth_aspect_ratio(mouth):
    # 计算距离，竖直的
    D = dist.euclidean(mouth[3], mouth[18])
    E = dist.euclidean(mouth[14], mouth[9])
    # 计算距离，水平的
    F = dist.euclidean(mouth[0], mouth[6])
    # ear值
    mou = (D + E) / (2.0 * F)
    return mou 

#关键点定位
def shape_to_np(shape, dtype="int"):
    # 创建68*2
    coords = np.zeros((shape.num_parts, 2), dtype=dtype)
    # 遍历每一个关键点
    # 得到坐标
    for i in range(0, shape.num_parts):
        coords[i] = (shape.part(i).x, shape.part(i).y)
    return coords

# 设置判断参数（需要调试）
EYE_AR_THRESH = 0.3  # ear小于0.3判断为闭眼
EYE_AR_CONSEC_FRAMES = 3  # 连续三帧ear都小于0.3判断为眨眼
MOU_AR_THRESH = 0.5  # mou大于0.5判断为张大嘴巴
MOU_AR_CONSEC_FRAMES = 10  # 连续十帧mou都大于0.5判断为眨眼

# 初始化计数器
COUNTER_ear = 0
COUNTER_mou = 0
TOTAL_ear = 0
TOTAL_mou = 0
start_time = time.time()

#加载dlib库中的人脸检测与关键点定位
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('.\shape_predictor_68_face_landmarks.dat')

#分别取两个眼睛区域
(lStart, lEnd) = FACIAL_LANDMARKS_68_IDXS["left_eye"]
(rStart, rEnd) = FACIAL_LANDMARKS_68_IDXS["right_eye"]
(mStart, mEnd) = FACIAL_LANDMARKS_68_IDXS["mouth"]

#以下不断循环检测，输出
while True:
    
    # 判断是否达到30秒时间间隔
    current_time = time.time()
    elapsed_time = current_time - start_time
    if elapsed_time >= 30:
        # 将数据插入MySQL数据库
        add_data(cursor,current_time,TOTAL_ear,TOTAL_mou)
        # 提交数据库事务
        connection.commit()

        # 重置计数器和时间变量
        TOTAL_ear = 0
        TOTAL_mou = 0
        start_time = time.time()
    
    # 读取视频帧
    ret, frame = cap.read()
    frame = cv2.flip(frame, 0)
    
    # 检查是否成功读取帧
    if not ret:
        print("无法获取视频帧")
        break
        
    (h, w) = frame.shape[:2]
    width=1200
    r = width / float(w)
    dim = (width, int(h * r))
    frame = cv2.resize(frame, dim, interpolation=cv2.INTER_AREA)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    
    # 检测人脸
    rects = detector(gray, 0)

    # 遍历每一个检测到的人脸
    for rect in rects:
        # 获取坐标
        shape = predictor(gray, rect)
        shape = shape_to_np(shape)

        # 分别计算ear值,mou值
        leftEye = shape[lStart:lEnd]
        rightEye = shape[rStart:rEnd]
        mouth = shape[mStart:mEnd]
        leftEAR = eye_aspect_ratio(leftEye)
        rightEAR = eye_aspect_ratio(rightEye)
        mou = mouth_aspect_ratio(mouth)

        # 算一个平均的
        ear = (leftEAR + rightEAR) / 2.0

        # 绘制眼睛区域
        leftEyeHull = cv2.convexHull(leftEye)
        rightEyeHull = cv2.convexHull(rightEye)
        cv2.drawContours(frame, [leftEyeHull], -1, (0, 255, 0), 1)
        cv2.drawContours(frame, [rightEyeHull], -1, (0, 255, 0), 1)

        #绘制嘴巴区域
        mouthHull = cv2.convexHull(mouth)
        cv2.drawContours(frame, [mouth], -1, (0, 255, 0), 1)     
        
        # 检查ear是否满足阈值
        if ear < EYE_AR_THRESH:
            COUNTER_ear += 1

        else:
            # 如果连续几帧都是闭眼的，总数算一次
            if COUNTER_ear >= EYE_AR_CONSEC_FRAMES:
                TOTAL_ear += 1

            # 重置
            COUNTER_ear = 0

        # 检查mou是否满足阈值
        if mou > MOU_AR_THRESH:
            COUNTER_mou += 1

        else:
            # 如果连续几帧都是闭眼的，总数算一次
            if COUNTER_mou >= MOU_AR_CONSEC_FRAMES:
                TOTAL_mou += 1

            # 重置
            COUNTER_mou = 0            
            
        # 显示
        cv2.putText(frame, "Blinks: {}".format(TOTAL_ear), (10, 30),
            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        cv2.putText(frame, "EAR: {:.2f}".format(ear), (300, 30),
            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        cv2.putText(frame, "Yawn: {}".format(TOTAL_mou), (10, 60),
            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        cv2.putText(frame, "MOU: {:.2f}".format(mou), (300, 60),
            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)        

    cv2.imshow("Frame", frame)
    key = cv2.waitKey(10) & 0xFF
    
    # 按下 'q' 键退出循环
    if cv2.waitKey(1) == ord('q'):
        # 关闭游标和数据库连接
        #cursor.close()
        #db_connection.close()
        break

# 释放视频捕获对象和窗口
cap.release()
cv2.destroyAllWindows()
```
